```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
orders=pd.read_csv("Orders.csv")
orders
```

```python
orders.info()
```

```python
orders["OrderDateTime"]=pd.to_datetime(orders["OrderDateTime"])
orders["ExpectedDeliveryTime"]=pd.to_datetime(orders["ExpectedDeliveryTime"])
orders["ActualDeliveryTime"]=pd.to_datetime(orders["ActualDeliveryTime"])
orders.info()
```

```python
#dt is for  datetime which is in module in python that can handlke date and times
orders["Act_dehlivery_in_mins"]=(orders["ActualDeliveryTime"] - orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
#expected delivery in minuits
orders["total Expected mins"]=(orders["ExpectedDeliveryTime"] - orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
orders["Act_dehlivery_in_mins"].unique() #distinct values in a column
```

```python
orders["diff_btw_exp_and_act"]=orders["total Expected mins"] - orders["Act_dehlivery_in_mins"]
orders
```

```python
orders.loc[(orders["diff_btw_exp_and_act"])==0,"order_status"] = "On Time"
orders.loc[(orders["diff_btw_exp_and_act"])< 0,"order_status"] = "Delayed"
orders
```

```python
sns.countplot(orders["order_status"])
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x="StoreID",data=orders,hue="order_status")# hue allows us to provide legend
plt.xticks(rotation=90)
plt.show()
```

```python
product=pd.read_csv("Products.csv")
product.head()
```

```python
#joining two tables
orders.head()
```

```python
df=pd.merge(left=orders,right=product,on="ProductID",how="inner")#use left for all the columns
df.head()
```

```python
sns.barplot(x=df["Category"],y=df["TotalAmount"],data=df,estimator="sum",palette="viridis")
plt.xticks(rotation=90)
plt.show()
```
